## **Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?**

```python
new_imdb_df.head()
```

```python
# creat the dataframe of non null crew values
non_null_crew=new_imdb_df[~new_imdb_df['crew'].isna()]
```

```python
non_null_crew.head(2)
```

```python
non_null_crew.loc[0,'crew'][0]
```

```python
# create function to get the list of producer in non_null_crew
def create_producer_list(index):
  movie_index=non_null_crew.iloc[index]
  crew_data=movie_index.loc['crew']
  producer_lists=[]
  for elem in crew_data:
    if elem['job']=='Producer':
      producer_lists.append(elem['name'])
  return producer_lists
```

```python
create_producer_list(61)
```

```python
# same way create the list of director from non null crew dataframe
def create_director_list(index):
  movie_index=non_null_crew.iloc[index]
  crew_data=movie_index.loc['crew']
  directors_lists=[]
  for elem in crew_data:
    if elem['job']=='Director':
      directors_lists.append(elem['name'])
  return directors_lists
```

```python
create_director_list(61)
```

```python
new_imdb_df.head(2)
```

```python
# creating the dataframe with the required columns and append the data later
table=pd.DataFrame(columns=['Movie title','Directors','producers','ROI'])
```

Appending the tabel Dadaframe with try and except block to bypass error because some values of the dictionaries contain float as value 

```python
for index,row in non_null_crew.iterrows():
  try:
    table=table.append({'Movie title':non_null_crew.loc[index,'original_title'],'producers':create_producer_list(index),'Directors':create_director_list(index),'ROI':non_null_crew.loc[index,'roi']},ignore_index=True)
  except:
    continue
```

```python
# table containing the columns the movie , directors and producer of the and movie and roi as well lets check it
table.head()
```

```python
def list_to_str(li):
 
    # initialize an empty string
    str1 = ""
 
    # traverse in the string
    for ele in li:
        str1 += ele
 
    # return string
    return str1
```

```python
table['Directors']=table['Directors'].apply(list_to_str)
table['producers']=table['producers'].apply(list_to_str)
```

```python
table.head()
```

```python
 table_df=table.groupby('producers')['ROI'].mean()
```

## **These are the top three producers which have highest average ROI**

```python
table_df.reset_index().sort_values(by='ROI',ascending=False).head(3)
```

```python
# find out the profit and movie of the corresponding actor in the movie
profit1=[]
profit2=[]
movie1=[]
movie2=[]
  

for index, row in non_null_cast.iterrows():
  for iter in non_null_cast.loc[index,'cast']:
    if type(iter)  == dict:
      actor=iter['name']
      if 'Samuel L. Jackson' in actor:
        profit1.append(non_null_cast.loc[index,'profits'])
        movie1.append(non_null_cast.loc[index,'original_title'])


      if 'Robert De Niro' in actor:
        profit2.append(non_null_cast.loc[index,'profits'])
        movie2.append(non_null_cast.loc[index,'original_title'])

```

```python
# keep this in array
genre_s=np.array(genre_samuel)
genre_r=np.array(genres_robert)
```
